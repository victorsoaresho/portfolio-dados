```python
# Importando as bibliotecas
import numpy as np
import pandas as pd
import os

import tensorflow as tf
import tensorflow_decision_forests as tfdf
```

```python
# Importando os dados
train_df = pd.read_csv('/content/train.csv')
serving_df = pd.read_csv('/content/test.csv')

train_df.head(10)
```

#Preparar conjunto de dados
##Aplicaremos as seguintes transformações no conjunto de dados.

1.Tokenizar os nomes. Por exemplo, "Braund, Mr. Owen Harris" se tornará ["Braund", "Mr.", "Owen", "Harris"].

2.Extrair qualquer prefixo no bilhete. Por exemplo, o bilhete "STON/O2. 3101282" se tornará "STON/O2." e 3101282.

```python
def preprocess(df):
    df = df.copy()

    def normalize_name(x):
        return " ".join([v.strip(",()[].\"'") for v in x.split(" ")])

    def ticket_number(x):
        return x.split(" ")[-1]

    def ticket_item(x):
        items = x.split(" ")
        if len(items) == 1:
            return "NONE"
        return "_".join(items[0:-1])

    df["Name"] = df["Name"].apply(normalize_name)
    df["Ticket_number"] = df["Ticket"].apply(ticket_number)
    df["Ticket_item"] = df["Ticket"].apply(ticket_item)
    return df

preprocessed_train_df = preprocess(train_df)
preprocessed_serving_df = preprocess(serving_df)

preprocessed_train_df.head(5)
```

Vamos manter a lista das características de entrada do modelo. Notavelmente, não queremos treinar nosso modelo com as características "PassengerId" e "Ticket".

```python
input_features = list(preprocessed_train_df.columns)
input_features.remove("Ticket")
input_features.remove("PassengerId")
input_features.remove("Survived")
#input_features.remove("Ticket_number")

print(f"Input features: {input_features}")
```

#Converter conjunto de dados Pandas para conjunto de dados TensorFlow

```python
def tokenize_names(features, labels=None):
    """Divida os nomes em tokens. O TF-DF pode consumir tokens de texto nativamente."""
    features["Name"] =  tf.strings.split(features["Name"])
    return features, labels

train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(preprocessed_train_df,label="Survived").map(tokenize_names)
serving_ds = tfdf.keras.pd_dataframe_to_tf_dataset(preprocessed_serving_df).map(tokenize_names)
```

#Treinar modelo com parâmetros padrão
##Treinar modelo
Primeiro, estamos treinando um modelo GradientBoostedTreesModel com os parâmetros padrão.

```python
model = tfdf.keras.GradientBoostedTreesModel(
    verbose=0, # Poucos logs
    features=[tfdf.keras.FeatureUsage(name=n) for n in input_features],
    exclude_non_specified_features=True, # Usar apenas as features em "features"
    random_seed=1234,
)
model.fit(train_ds)

self_evaluation = model.make_inspector().evaluation()
print(f"Acurácia: {self_evaluation.accuracy} Perda:{self_evaluation.loss}")
```

Treinar modelo com parâmetros padrão aprimorados

Agora você usará alguns parâmetros específicos ao criar o modelo GBT.

```python
model = tfdf.keras.GradientBoostedTreesModel(
    verbose=0, # Poucos logs
    features=[tfdf.keras.FeatureUsage(name=n) for n in input_features],
    exclude_non_specified_features=True, # Usar apenas as features em "features"

    #num_trees=2000,

    # Apenas para GBT.
    # Um pouco mais lento, mas ótimo para entender o modelo.
    # compute_permutation_variable_importance=True,

    # Alterar os hiperparâmetros padrão
    # hyperparameter_template="benchmark_rank1@v1",

    #num_trees=1000,
    #tuner=tuner

    min_examples=1,
    categorical_algorithm="RANDOM",
    #max_depth=4,
    shrinkage=0.05,
    #num_candidate_attributes_ratio=0.2,
    split_axis="SPARSE_OBLIQUE",
    sparse_oblique_normalization="MIN_MAX",
    sparse_oblique_num_projections_exponent=2.0,
    num_trees=2000,
    #validation_ratio=0.0,
    random_seed=1234,

)
model.fit(train_ds)

self_evaluation = model.make_inspector().evaluation()
print(f"Acurácia: {self_evaluation.accuracy} Perda:{self_evaluation.loss}")
```

Vamos analisar o modelo e você também pode notar as informações sobre a importância das variáveis que o modelo descobriu.

```python
model.summary()
```

# Fazendo previsões

```python
def prediction_to_kaggle_format(model, threshold=0.5):
    proba_survive = model.predict(serving_ds, verbose=0)[:,0]
    return pd.DataFrame({
        "PassengerId": serving_df["PassengerId"],
        "Survived": (proba_survive >= threshold).astype(int)
    })

def make_submission(kaggle_predictions):
    path="submission.csv"
    kaggle_predictions.to_csv(path, index=False)
    print(f"Submission exported to {path}")

kaggle_predictions = prediction_to_kaggle_format(model)
make_submission(kaggle_predictions)
```

#Treinando um modelo com ajuste de hiperparâmetros

O ajuste de hiperparâmetros é habilitado ao especificar o argumento do construtor do sintonizador do modelo. O objeto do sintonizador contém toda a configuração do sintonizador (espaço de busca, otimizador, teste e objetivo).

```python
tuner = tfdf.tuner.RandomSearch(num_trials=1000)
tuner.choice("min_examples", [2, 5, 7, 10])
tuner.choice("categorical_algorithm", ["CART", "RANDOM"])

local_search_space = tuner.choice("growing_strategy", ["LOCAL"])
local_search_space.choice("max_depth", [3, 4, 5, 6, 8])

global_search_space = tuner.choice("growing_strategy", ["BEST_FIRST_GLOBAL"], merge=True)
global_search_space.choice("max_num_nodes", [16, 32, 64, 128, 256])

#tuner.choice("use_hessian_gain", [True, False])
tuner.choice("shrinkage", [0.02, 0.05, 0.10, 0.15])
tuner.choice("num_candidate_attributes_ratio", [0.2, 0.5, 0.9, 1.0])


tuner.choice("split_axis", ["AXIS_ALIGNED"])
oblique_space = tuner.choice("split_axis", ["SPARSE_OBLIQUE"], merge=True)
oblique_space.choice("sparse_oblique_normalization",
                     ["NONE", "STANDARD_DEVIATION", "MIN_MAX"])
oblique_space.choice("sparse_oblique_weights", ["BINARY", "CONTINUOUS"])
oblique_space.choice("sparse_oblique_num_projections_exponent", [1.0, 1.5])

# Ajustar o modelo. Observe o `tuner=tuner`.
tuned_model = tfdf.keras.GradientBoostedTreesModel(tuner=tuner)
tuned_model.fit(train_ds, verbose=0)

tuned_self_evaluation = tuned_model.make_inspector().evaluation()
print(f"Accuracy: {tuned_self_evaluation.accuracy} Loss:{tuned_self_evaluation.loss}")
```

#Criando um conjunto (ensemble)

Aqui você irá criar 100 modelos com diferentes sementes e combinar seus resultados.

Esta abordagem remove um pouco os aspectos aleatórios relacionados à criação de modelos de ML.

Na criação do GBT, é utilizado o parâmetro honest. Ele usará diferentes exemplos de treinamento para inferir a estrutura e os valores dos nós folha. Esta técnica de regularização troca exemplos por estimativas de viés.

```python
predictions = None
num_predictions = 0

for i in range(100):
    print(f"i:{i}")
    # Modelos possíveis: GradientBoostedTreesModel ou RandomForestModel
    model = tfdf.keras.GradientBoostedTreesModel(
        verbose=0, # Poucos logs
        features=[tfdf.keras.FeatureUsage(name=n) for n in input_features],
        exclude_non_specified_features=True, ## Usar apenas as features em "features"

        #min_examples=1,
        #categorical_algorithm="RANDOM",
        ##max_depth=4,
        #shrinkage=0.05,
        ##num_candidate_attributes_ratio=0.2,
        #split_axis="SPARSE_OBLIQUE",
        #sparse_oblique_normalization="MIN_MAX",
        #sparse_oblique_num_projections_exponent=2.0,
        #num_trees=2000,
        ##validation_ratio=0.0,
        random_seed=i,
        honest=True,
    )
    model.fit(train_ds)

    sub_predictions = model.predict(serving_ds, verbose=0)[:,0]
    if predictions is None:
        predictions = sub_predictions
    else:
        predictions += sub_predictions
    num_predictions += 1

predictions/=num_predictions

kaggle_predictions = pd.DataFrame({
        "PassengerId": serving_df["PassengerId"],
        "Survived": (predictions >= 0.5).astype(int)
    })

make_submission(kaggle_predictions)
```
